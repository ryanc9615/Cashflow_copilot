```python
import pandas as pd
import random
from datetime import datetime, timedelta
from pathlib import Path
import duckdb
```

```python
categories = {
    "software": ["OPEN API", "NOTION LABS", "AWS CLOUD", "GITHUB"],
    "travel": ["UBER TRIP", "TRAINLINE UK", "RYANAIR FLIGHT"],
    "rent": ["WEWORK RENT", "OFFICE RENT LTD"],
    "marketing": ["GOOGLE ADS", "FACEBOOK ADS"],
    "meals": ["PRET A MANGER", "STARBUCKS"],
    "equipment": ["AMAZON", "APPLE STORE"]
}
```

```python
rows = []
```

```python
start_date = datetime(2025,1,1)
```

```python
category = random.choice(list(categories.keys()))
print(category)
```

```python
description = random.choice(categories[category])
print(description)
```

```python
amount = round(random.uniform(5,200),2)
print(amount)
```

```python
rows.append({
    "date": start_date + timedelta(days=random.randint(0,180)),
    "description": description,
    "amount": -amount,
    "currency": "GBP",
    "channel": random.choice(["card","transfer","direct_debit"]),
    "category": category
})

print(rows)
```

```python
df = pd.DataFrame(rows)
print(df)
```

```python
merch_key = df["merchant_key"] = df["description"].str.lower().str.split().str[0]
print(merch_key)
```
